```python

white_df = data.loc[(data['winner']=='white')]
white_win_count = white_df.groupby('opening_eco').size()
black_df = data.loc[(data['winner']=='black')]
black_win_count = black_df.groupby('opening_eco').size()

total_win = data.groupby('opening_eco').size()

total_win.reset_index()

win_white_percentage = ((white_win_count/total_win)* 100).fillna(0)

win_black_percentage = ((black_win_count/total_win)* 100).fillna(0)

win_percentage = pd.DataFrame({'white wins %': win_white_percentage, 'black wins %':win_black_percentage}).reset_index()

win_percentage.plot(x='opening_eco', kind= 'bar', figsize=(50,6))
plt.xticks(rotation=90)
plt.show()
```

```python
#22
import pandas as pd
import matplotlib.pyplot as plt

# Filter data for white and black winners
white_df = data.loc[data['winner'] == 'white']
black_df = data.loc[data['winner'] == 'black']

# Count wins for white and black players by opening_eco
white_win_count = white_df.groupby('opening_eco').size()
black_win_count = black_df.groupby('opening_eco').size()

# Calculate total games for each opening_eco
total_games = data['opening_eco'].value_counts()

# Calculate win percentages for white and black
win_white_percentage = (white_win_count / total_games * 100).fillna(0)
win_black_percentage = (black_win_count / total_games * 100).fillna(0)

print(win_white_percentage, win_black_percentage)
# Create a DataFrame for the win percentages
win_percentage = pd.DataFrame({
    'opening_eco': total_games.index,
    'white wins %': win_white_percentage,
    'black wins %': win_black_percentage
}).reset_index(drop=True)

# Plot the win percentages
win_percentage.plot(x='opening_eco', kind='bar', figsize=(200, 80), width=0.8, color=['blue', 'red'])

# Add labels and title
plt.title('Win Percentage for White and Black Players by Opening ECO', fontsize=160)
plt.xlabel('Opening ECO', fontsize=12)
plt.ylabel('Win Percentage (%)', fontsize=12)
plt.xticks(rotation = 90)
plt.legend(title='Player', fontsize=12)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
#23
data['rating_diff'] = data.apply(lambda row: row['white_rating'] - row['black_rating'], axis =1)
data.head()
```

```python
#24
data['game_length_category'] = data['turns'].apply(lambda turns:'Short' if turns<20 else 'Medium' if turns<=40 else 'Long')
data.head()
```

```python
#25
opening = data['opening_eco'].value_counts()
data['Opening popularity'] = data['opening_eco'].apply(lambda x:'Common' if opening[x]>100 else 'Rare')
data.head()
```

```python
# Calculate total games played by each player (white and black combined)
player_games = data['white_id'].value_counts() + data['black_id'].value_counts()

# Calculate total wins for each player
player_wins = data.loc[data['winner'] == 'white', 'white_id'].value_counts() + \
              data.loc[data['winner'] == 'black', 'black_id'].value_counts()

# Calculate win percentage for each player
player_win_percentage = (player_wins / player_games * 100).fillna(0)

# Create a mapping for performance categorization
performance_mapping = player_win_percentage.apply(
    lambda win_pct: 'High Performance' if win_pct > 60 else 'Low Performance'
)

# Map performance back to the dataset for both white and black players
data['white_performance'] = data['white_id'].map(performance_mapping)
data['black_performance'] = data['black_id'].map(performance_mapping)

# Display the first few rows to verify
#print(data[['white_id', 'white_performance', 'black_id', 'black_performance']].head())
data
```
