```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('HumanResource.csv')
```

```python
df
```

```python
sns.scatterplot(x=df['Salary'], y=df['EngagementSurvey'])
```

```python
sns.histplot(x=df['Salary'], kde=True)
```

```python
sns.kdeplot(x=df['EngagementSurvey'])
```

```python
sns.boxplot(y=df['Salary'], x=df['Sex'])
```

```python
sns.violinplot(x=df['Sex'], y=df['EngagementSurvey'])
```

```python
sns.jointplot(x=df['Salary'], y=df['EngagementSurvey'])
```

```python
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
sns.pairplot(df[numerical_cols])
plt.suptitle('Pair Plot of Numerical Variables', y=1.02)
plt.show()
```

```python
# Heatmap: Correlation Matrix
plt.figure(figsize=(10, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Heatmap: Correlation Matrix')
plt.show()
```
