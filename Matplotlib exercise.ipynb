```python
fig, ax = plt.subplots()
ax.scatter(x=df['Age'], y=df['Glucose'])
```

```python
fig, ax = plt.subplots()
ax.scatter(x=df['BMI'], y=df['Insulin'], s=df['Insulin']*0.5, color = "Green")
```

```python
# Count the occurrences of each outcome
    
diabetes_cases = df['Outcome'].value_counts()

# Convert to a dictionary for easier handling
diabetes_cases_dict = {
    "Negative": diabetes_cases.get(0, 0),  # 0 for negative cases
    "Positive": diabetes_cases.get(1, 0)   # 1 for positive cases
}
print(diabetes_cases_dict)
# Create a bar chart
plt.bar(diabetes_cases_dict.keys(), diabetes_cases_dict.values(), color=['blue', 'red'])

# Add labels and title
plt.xlabel("Diabetes Outcome")
plt.ylabel("Frequency")
plt.title("Frequency of Diabetes Cases (Positive/Negative)")

# Display the chart
plt.show()

```

```python
sizes = [df['Outcome'].value_counts()[0], df['Outcome'].value_counts()[1]]

# Create the pie chart
fig, ax = plt.subplots()
ax.pie(sizes, labels=['Outcome 0', 'Outcome 1'], autopct='%1.1f%%')
```
